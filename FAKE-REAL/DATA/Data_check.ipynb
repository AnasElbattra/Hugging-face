```python
import pandas as pd
import numpy as np
```

```python
erfan_train = pd.read_csv("ErfanMoosaviMonazzah/train.csv")
erfan_test = pd.read_csv("ErfanMoosaviMonazzah/test.csv")
gonzaloa_train = pd.read_csv("GonzaloA_fake_news/train Data.csv")
gonzaloa_test = pd.read_csv("GonzaloA_fake_news/test Data.csv")
ifnd = pd.read_csv("IFND/IFND.csv", encoding="latin-1")
kaggle_train = pd.read_csv("kaggle_datafake/kaggle_datafake_train.csv")
kaggle_test = pd.read_csv("kaggle_datafake/kaggle_datafake_test.csv")
kaggle_comp_train = pd.read_csv("Kaggle-Competition/train.csv")
kaggle_comp_test = pd.read_csv("Kaggle-Competition/test.csv")
mrmm = pd.read_csv("mrm8488-fake-news/fake_news.csv")
pushdeepfake_train = pd.read_csv("pushpdeepfake_news_combined/train.csv")
pushdeepfake_test = pd.read_csv("pushpdeepfake_news_combined/test.csv")
```

```python
ifnd = ifnd.rename(columns={"Label": "label", "Statement": "text"})
ifnd["label"] = ifnd["label"].map({"Fake": 0, "TRUE": 1})
```

```python
erfan_train = erfan_train[["text", "label"]]
erfan_train["df"] = "erfan_train"
erfan_test = erfan_test[["text", "label"]]
erfan_test["df"] = "erfan_test"
gonzaloa_train = gonzaloa_train[["text", "label"]]
gonzaloa_train["df"] = "gonzaloa_train"
gonzaloa_test = gonzaloa_test[["text", "label"]]
gonzaloa_test["df"] = "gonzaloa_test"
ifnd = ifnd[["text", "label"]]
ifnd["df"] = "ifnd"
kaggle_train = kaggle_train[["text", "label"]]
kaggle_train["df"] = "kaggle_train"
kaggle_test = kaggle_test[["text", "label"]]
kaggle_test["df"] = "kaggle_test"
kaggle_comp_train = kaggle_comp_train[["text", "label"]]
kaggle_comp_train["df"] = "kaggle_comp_train"
kaggle_comp_test = kaggle_comp_test[["text", "label"]]
kaggle_comp_test["df"] = "kaggle_comp_test"
mrmm = mrmm[["text", "label"]]
mrmm["df"] = "mrmm"
pushdeepfake_train = pushdeepfake_train[["text", "label"]]
pushdeepfake_train["df"] = "pushdeepfake_train"
pushdeepfake_test = pushdeepfake_test[["text", "label"]]
pushdeepfake_test["df"] = "pushdeepfake_test"
```

```python
# concat all dataframes and reset index
df = pd.concat(
    [
        erfan_train,
        erfan_test,
        gonzaloa_train,
        gonzaloa_test,
        ifnd,
        kaggle_train,
        kaggle_test,
        kaggle_comp_train,
        kaggle_comp_test,
        mrmm,
        pushdeepfake_train,
        pushdeepfake_test,
    ],
    ignore_index=True,
)
df
```
